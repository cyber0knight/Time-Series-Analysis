```python
def correlation_coefficient_cal(x,y):
    x_avg = sum(x) / len(x)
    y_avg = sum(y) / len(y)
    diff_x = x - x_avg
    diff_y = y - y_avg
    numerator = sum(diff_x * diff_y)
    denominator = round(sqrt(sum(diff_x ** 2)) * sqrt(sum(diff_y ** 2)), 2)
    r = numerator / denominator
    return r

def autocorrelation_cal(ts):
    avg = ts.mean()
    deviation_from_mean = ts - avg
    denominator = np.sum(deviation_from_mean**2)
    
    T = len(ts)
    acf_all_lags = []

    for k in range(0, T):       # k is # of lags
        t = k+1
        numerator = 0
        for i in range(t, T+1):
            numerator += (ts[i-1]-avg)*(ts[i-1-k]-avg)       # -1 is added in original equation bc python indexing starts at 0 instead of 1
        acf = numerator/denominator
        acf_all_lags.append(acf)
        # if (k == 100):
        #     break


    return pd.Series(acf_all_lags)


# def autocorrelation_cal_k_lags(ts, lags):
#     avg = ts.mean()
#     deviation_from_mean = ts - avg
#     denominator = np.sum(deviation_from_mean**2)

#     T = len(ts)
#     acf_all_lags = []

#     for k in range(0, T):       # k is # of lags
#         t = k+1
#         numerator = 0
#         for i in range(t, T+1):
#             numerator += (ts[i-1]-avg)*(ts[i-1-k]-avg)       # -1 is added in original equation bc python indexing starts at 0 instead of 1
#         acf = numerator/denominator
#         acf_all_lags.append(acf)
#         if (k == lags):
#             break

#     return pd.Series(acf_all_lags)

def autocorrelation_cal_k_lags(ts, lags):
    T = len(ts)
    avg = ts.mean()
    denominator = ((ts - avg) ** 2).sum()
    acf_all_lags = []
    for k in range(1, lags + 1):
        numerator = sum((ts.iloc[i-1] - avg) * (ts.iloc[i-1-k] - avg) for i in range(k, T+1))
        # numerator = sum((ts[i-1] - avg) * (ts[i-1-k] - avg) for i in range(k, T+1))
        acf = numerator / denominator
        acf_all_lags.append(acf)
    return acf_all_lags
```

ADF Test in Time Series
The Augmented Dickey-Fuller (ADF) test is a statistical technique used to determine whether a time series is stationary or not. Stationarity is a crucial assumption in many time series models, including ARIMA, Exponential Smoothing, and Vector Autoregression.

What is Stationarity?
A stationary time series is one that has a constant mean, variance, and autocorrelation structure over time. In other words, the series looks similar when plotted over time, with no trends, seasonality, or sudden changes.

What is the ADF Test?
The ADF test is an extension of the Dickey-Fuller (DF) test, which was originally designed to test for unit roots in a time series. A unit root is a special case of non-stationarity, where the time series has a stochastic trend that persists over time.

The ADF test is based on the idea of testing whether the coefficient of the lagged level of the time series is equal to one in a regression equation. The null hypothesis of the test is that the time series has a unit root (i.e., it is non-stationary), and the alternative hypothesis is that it is stationary.

How to Interpret ADF Test Results
Test Statistic: The ADF test produces a test statistic, which is a measure of how strongly the data rejects the null hypothesis of a unit root.
p-Value: The p-value associated with the test statistic indicates the probability of observing the test statistic (or a more extreme value) under the null hypothesis of a unit root. A small p-value (typically < 0.05) suggests that the null hypothesis can be rejected, indicating that the time series is likely stationary.
Interpretation: If the null hypothesis is rejected, it implies that the time series is stationary or trend-stationary. If the null hypothesis cannot be rejected, it suggests that the time series may have a unit root and is therefore non-stationary.
Common Applications of ADF Test
Time Series Modeling: Before building a time series model, it’s essential to determine whether the series is stationary. The ADF test helps identify non-stationary series, which require differencing or other transformations to achieve stationarity.
Spurious Regression: The ADF test can detect spurious regression, where two unrelated time series appear to be correlated due to their common trends.
Forecasting: Stationarity is a prerequisite for many forecasting techniques. The ADF test ensures that the time series is suitable for forecasting models.
In summary, the Augmented Dickey-Fuller (ADF) test is a statistical technique used to determine whether a time series is stationary or not. It’s essential for time series modeling, spurious regression detection, and forecasting.

```python
def ADF_test(df_col):
    result = adfuller(df_col)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
```

```python
def calc_ARMA(samples, coeff_MA, coeff_AR):
    np.random.seed(42)
    mean = 0; std = 1
    e = std * (np.random.randn(samples)+mean)
    system = (coeff_MA, coeff_AR, 1)
    y_dlsim = signal.dlsim(system, e)
    y = y_dlsim[1].flatten()        # # dlsim fn returns a tuple (the [1] element of which is our desired ARMA process but it is in 2d so flatten)

    return y
```

The function phi_kk(j, k, acf) calculates the phi coefficient (or the partial autocorrelation) for a given lag using the Cramer's Rule for solving determinants. This is typically used in autoregressive models (AR) to calculate the coefficients of the AR model.

Let's break down the code:

Arguments:
j: This is the lag for which you're calculating the partial autocorrelation (i.e., the point at which the correlation is being computed).
k: This is the order for the matrix, determining the size of the matrices used in the calculation (usually corresponding to the AR model's order).
acf: This is the autocorrelation function, typically a list or array containing the autocorrelation values at various lags.
Step-by-step Explanation:
Create a denominator matrix (den): The function first creates a square matrix den of size 
𝑘×𝑘 k×k. The entries of this matrix are filled with the values from the acf array at certain lag positions, specifically the absolute difference between the row and column indices (using acf[abs(j + row - col)]).

den[row, col] = acf[abs(j + row - col)]: This fills the matrix den by referencing the autocorrelation values at the lags based on the distance from the given lag 
𝑗
j.
Create a copy for the numerator matrix (num): A copy of the denominator matrix den is made, and the last column of num is replaced by the values of the acf array starting from j+1 to j+k+1. This is done to set up for Cramer's Rule.

num[:, -1] = acf[j+1:j+k+1]: Replaces the last column of num with the autocorrelation values at the lags from  𝑗 + 1 j+1 to 𝑗 + 𝑘 j+k.
Calculate the phi coefficient:

The phi coefficient is computed using Cramer's Rule: 𝜙 𝑘 𝑘 = det(num) det(den) ϕ  kk ​ =  det(den) det(num)​
 
This involves taking the determinant of both the numerator (num) and the denominator (den) matrices, and dividing the determinant of the numerator by the determinant of the denominator.

phi = round(np.linalg.det(num)/np.linalg.det(den), 3): Calculates the ratio of the determinants, rounded to 3 decimal places.
Return the value of phi: The computed phi value is returned.

Purpose:
This function is used in time series analysis, especially in autoregressive (AR) models, to compute the partial autocorrelation coefficient for a given lag. This coefficient helps in determining the strength of a relationship between a series and its lagged values, accounting for intermediate lags.

Example Usage:
Suppose you want to compute the partial autocorrelation for a specific lag, say j = 3 and for an AR(2) model (k = 2), using the autocorrelation values stored in acf. You would call the function as follows:

```python
def phi_kk(j,k,acf):

    den = np.zeros((k,k))   # denominator matrix for phi_kk
    for row in range(k):
        for col in range(k):
            den[row, col] = acf[abs(j+row-col)]

    num = den.copy()
    num[:, -1] = acf[j+1:j+k+1]     # replacing last column of num for cramer's rule

    phi = round(np.linalg.det(num)/np.linalg.det(den), 3)

    return phi
```

```python
acf = [1, 0.8, 0.5, 0.3, 0.1, 0.05, 0.02]  # Example ACF values
phi_value = phi_kk(3, 2, acf)
print(f"The phi coefficient for j=3 and k=2 is: {phi_value}")
```

# SIMPLE FORECASTING METHODS

```python
def naive_method_hstep(train):
    return train.iloc[-1]
```

```python
def moving_average(data,p=10):
    n = len(data)
    predictions = []
    currSum = 0 
    for i in range(p):
        currSum += data[i]
    for i in range(p,n):
        predictions.append(currSum/p)
        currSum -= data[i-p]
        currSum  += data[i]
    return predictions
```

```python
def exponential_smoothing(data , alpha = 0.95):
    base = np.mean(data)
    y = 0
    predictions = []
    for i in data:
        pred = alpha*base + (1-alpha)*y
        y = i 
        base = pred
        predictions.append(pred)
    return predictions[1:]
```

```python
def holt_linear(train, l):
    model = ets.ExponentialSmoothing(train, trend='mul', damped=True)     # Use the multiplicative version, unless the data has been logged before.
    fit1 = model.fit()
    print('Holt Linear method summary\n',fit1.summary())
    return fit1.predict(start=l.index[0], end=l.index[-1])
```

```python
def holt_winter_seasonal(train, l):
    # Specify the number of periods in a season, e.g., for monthly data with yearly seasonality, it's 12
    model = ets.ExponentialSmoothing(train, trend='mul', seasonal='mul', seasonal_periods=12)
    fit1 = model.fit()

    # Forecasting based on the test data index
    return fit1.predict(start=l.index[0], end=l.index[-1])
```

```python
def autocorrelation_cal_k_lags2(ts, lags):
    T = len(ts)
    avg = ts.mean()
    denominator = ((ts - avg) ** 2).sum()
    acf_all_lags = []
    for k in range(1, lags + 1):
        # numerator = sum((ts.iloc[i-1] - avg) * (ts.iloc[i-1-k] - avg) for i in range(k, T+1))
        numerator = sum((ts[i-1] - avg) * (ts[i-1-k] - avg) for i in range(k, T+1))
        acf = numerator / denominator
        acf_all_lags.append(acf)
    return acf_all_lags
```

```python
def ARMA_estimates(na, nb):
    train_arma, test_arma = train_test_split(sty_df.Open, test_size=0.2, shuffle=False)

    # Use SARIMAX to fit ARMA model
    arma_model = sm.tsa.SARIMAX(train_arma, order=(na, 0, nb), trend='n').fit(disp=False)

    for i in range(na):
        print(f'The AR coefficient a{i} is:', arma_model.params[i])
    for i in range(nb):
        print(f'The MA coefficient b{i} is:', arma_model.params[i + na])
    print()

    print(arma_model.summary())

    print('Confidence Interval (95%) for estimated parameters are:\n', arma_model.conf_int(alpha=0.05))
    print()

    # Estimated covariance matrix
    print('The estimated covariance matrix for the model is:\n', arma_model.cov_params())
    print()

    # Corrected Attribute Error for variance
    print('The estimated variance & standard deviation of prediction error:\n', arma_model.scale, np.sqrt(arma_model.scale))
    print()

    # Prediction
    model_hat = arma_model.predict(start=test_arma.index[0], end=test_arma.index[-1])

    # Plot original vs predictions
    empty_list = [None] * len(train_arma)
    plt.plot(sty_df.Open, label='Original (differenced) values')
    plt.plot(empty_list + list(model_hat), label='ARMA Predictions')
    plt.legend()
    plt.xlabel('Time')
    plt.xticks(sty_df.index[::48], rotation=45)
    plt.ylabel('Values obtained')
    plt.title(f'True vs Predicted values (ARMA({na},{nb}) Model)')
    plt.show()

    # Calculate residuals
    residuals_arma = pd.DataFrame(model_hat.values - test_arma.values)
    a = np.array(residuals_arma[0])
    a = np.delete(a, -1)

    print('Mean of ARMA residuals:', a.mean())
    print('Variance of ARMA residuals:', a.var())
    print('RMSE of ARMA residuals:', np.sqrt(np.mean(a**2)))

    # Calculate Q value
    acf_residuals_arma = autocorrelation_cal_k_lags2(a, 20)
    N = len(test_arma)
    Q_arma = N * (np.sum(np.array(acf_residuals_arma[1:])**2))

    DOF = 20 - na - nb
    alpha = 0.01
    Q_critical = chi2.ppf(1 - alpha, DOF)

    if Q_arma < Q_critical:
        print(f'Q ({Q_arma}) < Q_c ({Q_critical}) hence residuals pass the whiteness test.')
    else:
        print('Residuals fail Chi-Square test.')

    # Reverse transformation with dynamic index handling
    last_train_index = train_arma.index[-1]
    reverse_diff_arma_predictions = df.Open.iloc[last_train_index] + model_hat.cumsum()

    plt.plot(df.Open, label='True values')
    plt.plot(empty_list + list(reverse_diff_arma_predictions), label='ARMA Predictions')
    plt.legend()
    plt.xlabel('Time')
    plt.xticks(df.index[::48], rotation=45)
    plt.ylabel('Values obtained')
    plt.title(f'True vs Predicted values ARMA({na},{nb}) (After Reverse Transform)')
    plt.show()

    return residuals_arma, reverse_diff_arma_predictions

print('********* Results for ARMA(1,0) start here *************')
# residuals_arma, reverse_diff_arma_predictions = ARMA_estimates(1, 0)
```

```python
def create_gpac(residuals, max_p=10, max_q=10):
    gpac_table = np.zeros((max_p, max_q))
    
    for p in range(1, max_p + 1):
        for q in range(1, max_q + 1):
            num_matrix = np.zeros((p, p))
            denom_matrix = np.zeros((p, p))
            
            for i in range(p):
                for j in range(p):
                    index_denom = abs(q + i - j)
                    index_num = abs(q + i + 1)
                    
                    if index_denom < len(residuals):
                        denom_matrix[i, j] = residuals[index_denom]
                    else:
                        denom_matrix[i, j] = 0  # Handle out-of-bounds safely
                    
                    if j == p - 1 and index_num < len(residuals):
                        num_matrix[i, j] = residuals[index_num]
                    elif index_denom < len(residuals):
                        num_matrix[i, j] = residuals[index_denom]
                    else:
                        num_matrix[i, j] = 0  # Handle out-of-bounds safely
            
            numerator = np.linalg.det(num_matrix)
            denominator = np.linalg.det(denom_matrix)
            gpac_table[p - 1, q - 1] = numerator / denominator if denominator != 0 else 0
    
    return gpac_table

gpac_arma_residuals = create_gpac(np.array(residuals_arma[0]))

print('*********  Results for ARMA(10,0) start here *************')
ARMA_estimates(10,0)
print('*********  Results for ARMA(11,0) start here *************')
ARMA_estimates(11,0)
print('*********  Results for ARMA(16,0) start here *************')
ARMA_estimates(16,4)


### ARIMA MODEL
print('****************** Auto ARIMA Model results start here *************')
import pmdarima as pm

# fitting arima model
arima_model = pm.auto_arima(train.Open, test='adf')
print(arima_model.summary())
arima_model.plot_diagnostics(figsize=(8,8))
plt.show()

# predicting
arima_model_hat = arima_model.predict(test.shape[0])  # predict N steps into the future


# creating prediction plot
empty_list = [None]*len(train)
plt.plot(df.Open, label= 'Original values')
plt.plot(empty_list + list(arima_model_hat), label='ARIMA(4,1,5) Predictions')
plt.legend()
plt.xlabel('Time')
plt.xticks(sty_df.index[::48], rotation=45)
plt.ylabel('Values obtained')
plt.title('True vs Predicted values (ARIMA(2,1,2) Model)')
plt.show()


# ***********************************************************************************************
# 9- Plot the ACF of the residuals.
# ***********************************************************************************************

residuals_arima = pd.DataFrame( arima_model_hat - test.Open)
residuals_arima['Open'] = residuals_arima[0]
a = np.array(residuals_arima['Open'])
# print(a)
# a = np.delete(a, -1)

# print(len(residuals_arima))
# print(len(test))

print('Mean of ARIMA residuals', a.mean())
print('Variance of ARIMA residuals', a.var())
print('RMSE of ARIMA residuals', np.sqrt(np.mean(a**2)))


acf_residuals_arima = autocorrelation_cal_k_lags2(a, 20)      #  lag 20
title = 'Residuals ARIMA(2,1,2)'
# N = len(test_arma)
plot_ACF_title(acf_residuals_arima, title)

# ***********************************************************************************************
# 10 - Find Q value.
# ***********************************************************************************************
N = len(test)
Q_arima = N * (np.sum(np.array(acf_residuals_arima[1:])**2))

# ***********************************************************************************************
# 11 - Are the residuals errors white? Knowing the DOF and alfa = .01 perform a   test and check if the residuals pass the whiteness test.

# Q must be less than Qc
# ***********************************************************************************************

DOF = 20 - 2 - 2
# define probability
alpha = 0.01

# retrieve value <= probability
Q_critical = chi2.ppf(1-alpha, DOF)

if Q_arima < Q_critical:
    print('Q ({}) < Q_c ({}) hence residuals pass the whiteness test confirming that suggested model is good.'.format(Q_arima, Q_critical))
else:
    print('ARIMA Residuals fail Chi-Square test.')


# print('****************** Auto ARIMA Model with y-intercept removed *************')
# arima_model = pm.auto_arima(train.Open, test='adf', with_intercept=False)
# print(arima_model.summary())
# arima_model.plot_diagnostics(figsize=(8,8))
# plt.show()
```
